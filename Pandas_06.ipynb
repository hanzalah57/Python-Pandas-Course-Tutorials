```python
import pandas as pd
data = pd.read_csv('New_sales.csv')
```

```python
data
```

# Data Sorting 

```python
data.sort_values(by='Sales')
```

```python
#sort by two or more columns by descending order
data.sort_values(by=['Sales', 'Trade'], ascending=False)
```

```python
#sort whole data using index
data.sort_index(ascending=False)
```

```python
#Sort particular series 
data['Sales'].sort_values(ascending=True)
```
